### 1. Module import

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

### 2. Data input

```python
# seed 값 설정
np.random.seed(3)
tf.random.set_seed(3)

# 데이터 입력
df = pd.read_csv('C:/Users/82108\Downloads/deeplearning/dataset/iris.csv', 
                 names = ["sepal_length", "sepal_width", "petal_lengh", "petal_width", "species"])
df.index = df.index+1
print(df.head())
```

### 3. Pairplot graph for totaldata

```python
# 그래프
sns.pairplot(df, hue='species');
plt.show()
```

### 4. One-hot-encoding

```python
# 데이터 분류
#("Iris-setosa"등과 같이 데이터 안에 문자열이 포함되어 있기 때문에 numpy를 이용하는 것이 아닌 pandas 이용)
dataset = df.values
X = dataset[:,0:4].astype(float)
Y_obj = dataset[:,4]

# 문자열을 숫자로 변환
#(Y값이 숫자가 아니라 문자열이기 떄문에 LabelEncoder()를 이용해 클래스 이름을 숫자로 변환)
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
#(활성화함수 적용을 위해서는 Y값이 0과 1로만 이루어져 있어야 하므로 tf.keras.utils.categorical()함수로 변환)
Y_encoded = tf.keras.utils.to_categorical(Y)
```

### 5. Model setting and running

```python
# 모델의 설정
model = Sequential()
model.add(Dense(16, input_dim=4, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(3, activation='softmax'))

# 모델 컴파일
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델 실행
model.fit(X, Y_encoded, epochs=500, batch_size=5)

# 결과 출력
print("|n Accuracy: %.4f" % (model.evaluate(X, Y_encoded)[1]))
print()
print("Percentage of Accuracy is ==>  %.2f%s" % ((model.evaluate(X, Y_encoded)[1])*100, '%'))
```

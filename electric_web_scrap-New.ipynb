```python
"""
월마다 일수가 다른 것을 제외하고는 동작이 동일하기 때문에 함수처리를 하려 했으나 시험해본 결과, 
NoSuchElementException 예외처리를 하지 못하고 동작이 멈춰서 전체 코드가 길어 지더라도 달마다
다 입력을 하는 방법으로 결정.
"""

# 데이터 저장을 위한 csv 
import csv

# Selenium : 
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.keys import Keys

path = "C:/Users/Hwi/Downloads/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(path)

# 기상포털 주소 1차 접속. 데이터가 충분히 표시된 2021년 1월 1일을 시작 화면으로 설정
driver.get('https://www.wunderground.com/history/daily/gb/london/EGLC/date/2021-1-1')
time.sleep(3)

# 온도 표기 선택창 클릭
driver.find_element(By.XPATH, '//*[@id="wuSettings"]/i').click()
time.sleep(3)

# 섭씨 선택
driver.find_element(By.XPATH, '//*[@id="wuSettings-quick"]/div/a[2]').click()
time.sleep(3)

# YYYY
# offset 2012
y = 2013
while(y<2016):
    y=y+1
    print('<<',y,'년','>>')
    
    # MM
    m = 0
    while(m<12):
        m=m+1
        print('<',m,'월','>')
        
        # DD
        # 01
        if(m==1):
            d=0
            while(d<31):
                d=d+1
                print(d,'일')
                driver.get('https://www.wunderground.com/history/daily/gb/london/EGLC/date/'+str(y)+'-'+str(m)+'-'+str(d))
                time.sleep(5)
                
                h = 0 # 여기서는 하루를 30분 단위로 쪼갠 온도 개수, 데이터의 개수를 h로 설정해서 반복시켰는데, 48이 안되는 경우도 있음.
                while(h<48):
                    h=h+1
                    try:
                        clk = driver.find_element(By.XPATH,'//*[@id="inner-content"]/div[2]/div[1]/div[5]/div[1]/div/lib-city-history-observation/div/div[2]/table/tbody/tr['+str(h)+']/td[1]/span')
                        clk_str = clk.text
                        clk_str.split()
                        temp_list = [int(y), int(m), int(d)]
                        temp_list.append(clk_str)
                        temp_list.append(int(h))
                        temp = driver.find_element(By.XPATH,'//*[@id="inner-content"]/div[2]/div[1]/div[5]/div[1]/div/lib-city-history-observation/div/div[2]/table/tbody/tr['+str(h)+']/td[2]')
                        temp_str = temp.text
                        temp_list.append(temp_str)
                        time.sleep(0.1)
                        print(temp_list) # 확인용 print
                        # 시간 루프 종료 지점
                    
                        # csv 추가 및 저장
                        with open("./"+str(y)+'_'+str(m)+'_'+"London_Temperature.csv", 'a', newline='') as file:
                            writer = csv.writer(file)
                            writer.writerow(temp_list)
                                          
                    except NoSuchElementException:
                        print('데이터 48개 미달')
                        break

                print('Day End')
                print()
                time.sleep(3)
                
        # 02
        if(m==2):
            d=0
            while(d<28):
                d=d+1
                print(d,'일')
                
        # 03
        if(m==3):
            d=0
            while(d<31):
                d=d+1
                print(d,'일')
                
        # 04
        if(m==4):
            d=0
            while(d<30):
                d=d+1
                print(d,'일')
                
        # 05
        if(m==5):
            d=0
            while(d<31):
                d=d+1
                print(d,'일')
                
        # 06
        if(m==6):
            d=0
            while(d<30):
                d=d+1
                print(d,'일')
                
        # 07
        if(m==7):
            d=0
            while(d<31):
                d=d+1
                print(d,'일')
                
        # 08
        if(m==8):
            d=0
            while(d<31):
                d=d+1
                print(d,'일')
                
        # 09
        if(m==9):
            d=0
            while(d<30):
                d=d+1
                print(d,'일')
                
        # 10
        if(m==10):
            d=0
            while(d<31):
                d=d+1
                print(d,'일')
                
        # 11
        if(m==11):
            d=0
            while(d<30):
                d=d+1
                print(d,'일')
                      
        # 12
        if(m==12):
            d=0
            while(d<31):
                d=d+1
                print(d,'일')
                      
        print()
    print('\n')
    if (y == 2016):
        print('끝!!')
        break 
```

```python
    # 해당 월 클릭 (2013년 1월부터 2016년12월에 해당되는 110~63 선택)
    m = 111
    while(m > 61):
        m = m-1
        # 연도와 월 선택창 클릭
        driver.find_element(By.XPATH, '//*[@id="month"]/option[' + str(m) + ']').click()
        
        # 달마다 구성일수가 다르기 때문에 반복일자 구별을 위해 월 출력
        month = driver.find_element(By.XPATH, '//*[@id="wt-his-title"]')
        month_str = month.text
        month_list = month_str.split()
        month_5 = month_list[5]
        time.sleep(5)
        
        """
        각 달마다 날짜, 온도 표기는 동일한 과정이라 함수로 처리할려 했으나, 연도의 마지막에 모아서 
        저장하려면 각 달마다 리스트 저장 이름을 다르게 해서 마지막에 통합 저장해야 함.
        """
        
        # -----------------------------------------< 01 >-----------------------------------------------
        if (month_5 == 'January'):
            print('1월')
            # 1일~31일 크롤링
            d=0
            while(d<31):
                d=d+1
                driver.find_element(By.XPATH, '//*[@id="wt-his-select"]/option[' + str(d) + ']').click()
                time.sleep(3)
                
                # 1시~24시 온도 스크래핑
                for h in range(1, 25):
                    temp_str = driver.find_element(By.XPATH,
                                                   '//*[@id="wt-his"]/tbody/tr[' + str(h) + ']/td[2]').text
                    date = driver.find_element(By.XPATH, '//*[@id="wt-his-title"]')
                    date_str = date.text
                    date_list = date_str.split()
                    Jan_list = date_list[4:7]
                    Jan_list.reverse()
                    Jan_list.append((str(h)+'h'))
                    Jan_list.append(temp_str) # 계속 추가됨
                    print(Jan_list) # 확인용 print
```

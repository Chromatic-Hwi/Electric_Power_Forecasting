```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.timeanddate.com/weather/uk/london/historic?month=11&year=2012'

response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    print(soup)

else : 
    print(response.status_code)
```

```python
# 본격적으로 해봅시다
import requests
from bs4 import BeautifulSoup as bs

# Selenium : 
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.keys import Keys

path = "C:/Users/Hwi/Downloads/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(path)

# webdriver가 런던 기상 기록 페이지에 접속하도록 명령, 시작은 2022년으로 지정
driver.get('https://www.timeanddate.com/weather/uk/london/historic?month=11&year=2022')

d = 2012
while(d < 2017):
    d = d+1
    
    # 연도와 월 선택창 클릭
    driver.find_element(By.XPATH,'//*[@id="month"]').click()

    # 해당 월 클릭 (2013년 1월부터 2017년1월에 해당되는 109~61 선택)
    # 2016년은 윤년이라 2월 29일이 있지만 하루차이라 생략하기로 함.
    for i in range(110, 60, -1):
        driver.find_element(By.XPATH, '//*[@id="month"]/option[' + str(i) + ']').click()
        
        # 달마다 구성일수가 다르기 때문에 반복일자 구별을 위해 월 출력
        month = driver.find_element(By.XPATH, '//*[@id="wt-his-title"]')
        month_str = month.text
        month_list = month_str.split()
        month_5 = month_list[5]
        time.sleep(5)
        
        
        # 01
        if (month_5 == 'January'):
            # 1일~31일
            for i in range(1, 32):
                driver.find_element(By.XPATH, '//*[@id="wt-his-select"]/option[' + str(i) + ']').click()
                time.sleep(5)
                # 1시~24시
                for j in range(1, 25):
                    temp_str = driver.find_element(By.XPATH,
                                                   '//*[@id="wt-his"]/tbody/tr[' + str(j) + ']/td[2]').text
                    date = driver.find_element(By.XPATH, '//*[@id="wt-his-title"]')
                    date_str = date.text
                    date_list = date_str.split()
                    date = date_list[4:7]
                    print(date, temp_str)
                print()
                time.sleep(10)
            
        # 02
        if (month_5 == 'February'):
            for i in range(1, 29):
                driver.find_element(By.XPATH, '//*[@id="wt-his-select"]/option[' + str(i) + ']').click()
                time.sleep(3)
        
        # 03
        if (month_5 == 'March'):
            for i in range(1, 32):
                driver.find_element(By.XPATH, '//*[@id="wt-his-select"]/option[' + str(i) + ']').click()
                time.sleep(3)
        
        # 04
        if (month_5 == 'April'):
            for i in range(1, 31):
                driver.find_element(By.XPATH, '//*[@id="wt-his-select"]/option[' + str(i) + ']').click()
                time.sleep(3)
            
        # 05
        if (month_5 == 'May'):
            for i in range(1, 31):
                driver.find_element(By.XPATH, '//*[@id="wt-his-select"]/option[' + str(i) + ']').click()
                time.sleep(3)
                
        # 06
        if (month_5 == 'June'):
            for i in range(1, 31):
                driver.find_element(By.XPATH, '//*[@id="wt-his-select"]/option[' + str(i) + ']').click()
                time.sleep(3)
                
        # 07
        if (month_5 == 'July'):
            for i in range(1, 31):
                driver.find_element(By.XPATH, '//*[@id="wt-his-select"]/option[' + str(i) + ']').click()
                time.sleep(3)
                
        # 08
        if (month_5 == 'August'):
            for i in range(1, 31):
                driver.find_element(By.XPATH, '//*[@id="wt-his-select"]/option[' + str(i) + ']').click()
                time.sleep(3)
                
        # 09
        if (month_5 == 'September'):
            for i in range(1, 31):
                driver.find_element(By.XPATH, '//*[@id="wt-his-select"]/option[' + str(i) + ']').click()
                time.sleep(3)
                
        # 10
        if (month_5 == 'October'):
            for i in range(1, 31):
                driver.find_element(By.XPATH, '//*[@id="wt-his-select"]/option[' + str(i) + ']').click()
                time.sleep(3)
                
        # 11
        if (month_5 == 'November'):
            for i in range(1, 31):
                driver.find_element(By.XPATH, '//*[@id="wt-his-select"]/option[' + str(i) + ']').click()
                time.sleep(3)
                
        # 12
        if (month_5 == 'December'):
            for i in range(1, 31):
                driver.find_element(By.XPATH, '//*[@id="wt-his-select"]/option[' + str(i) + ']').click()
                time.sleep(3)
                
        
      
```

```python
import datetime
import pandas as pd
import numpy as np
import csv

def get_days(yyyy, mm, dd):
    days = ['월', '화', '수', '목', '금', '토', '일']
    return days[datetime.date(yyyy, mm, dd).weekday()]

"""
영국의 휴일로는 공통휴일로 선정.
새해 첫날, 성 금요일(=부활절 이틀 전 금요일), 노동절 휴일(5월 첫 월요일), 춘계 휴일(5월의 마지막 월요일), 크리스마스, 박싱데이(12/26)들과 대체휴일
"""

holidays = ['2013-1-1', '2013-3-29', '2013-5-6', '2013-5-27', '2013-12-25', '2013-12-26',
           '2014-1-1', '2014-4-18', '2014-5-5', '2014-5-26', '2014-12-25', '2014-12-26',
           '2015-1-1', '2015-4-3', '2015-5-4', '2015-5-25', '2015-12-25', '2015-12-26', '2015-12-28',
           '2016-1-1', '2016-3-25', '2016-5-2', '2016-5-30', '2016-12-25', '2016-12-26', '2016-12-27']

month = {1:31, 2:28, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}
day=''
holiday=''

for y in range(2013, 2017):
    
    for m in range(1,13):
        csv_for_days = pd.read_csv("./Electirc_Data_CSV_Null_Disposed/"+str(y)+'_'+str(m)+'_'+"House1_Ch1_Electric_Data_ND.csv", 
                                   names = ['DateTime', 'Seq', 'Watt'], encoding='cp949')
        s = 0
        
        for d in range(month[m]):
            d=d+1 
            days=csv_for_days['DateTime'][d*24-1]
            days2=days.split()
            yyyy=int(days2[0][:-1])
            mm=int(days2[1][:-1])
            dd=int(days2[2][:-1])
            ymd = '{}-{}-{}'.format(yyyy,mm,dd) # 휴일 필터링을 위한 검사용 변수
            day = get_days(yyyy, mm, dd) # 요일 변수
            
            one_day_list_p=[]
            one_day_list_d=[]
            
            # 조건부 표현식으로 간단하게 휴일 처리
            holiday='O' if ymd in holidays or day=='토' or day=='일' else 'X'
            
            for n in range(s, s+24):
                date = csv_for_days['DateTime'][n]
                days3=date.split()
                yyyy2=int(days3[0][:-1])
                mm2=int(days3[1][:-1])
                dd2=int(days3[2][:-1])
                YMD = '{}{}{}'.format(yyyy2,mm2,dd2)
                power = csv_for_days['Watt'][n]
                one_day_list_d.append(YMD)
                one_day_list_p.append(power)
                
            for n in range(24):
                    new_list=[]
                    new_list.append(one_day_list_d[n])
                    new_list.append(day)
                    new_list.append(holiday)
                    new_list.append(int(n))
                    new_list.append(one_day_list_p[n])
                    
                    print(new_list)
                    print()
                    with open("./Electirc_Data_CSV_Final2/"+str(y)+'/'+str(y)+'_'+str(m)+'_'+"House1_Ch1_Electric_Data_Final2.csv", 'a', newline='') as file:
                                writer = csv.writer(file)
                                writer.writerow(new_list)
                                
            s=s+24
            
print('\n'*4,'------------------------------< 요일 입력, 휴일 처리 끝!! >------------------------------\n')  
```

```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

Feb_13  = pd.read_csv("./Electirc_Data_CSV/2013_2_House1_Ch1_Electric_Data.csv", 
                  names = ['Time', 'Watt'], 
                  header = None, encoding='cp949')
```

```python
plt.figure(figsize=(20, 12))

x=np.arange(24)
plt.xticks(np.arange(24))

#plt.yticks(np.arange(0, 7000))
#plt.yticks([], [])

for i in range(7, 15):
    y = Feb_13['Watt'][i*24:(i+1)*24]
    plt.plot(x, y)

plt.title("Graph", fontsize=15)
plt.xlabel("Hour",fontsize=13)
plt.ylabel("Watt",fontsize=13)

plt.savefig('./Null_Data_Graph.png')
plt.show()
```

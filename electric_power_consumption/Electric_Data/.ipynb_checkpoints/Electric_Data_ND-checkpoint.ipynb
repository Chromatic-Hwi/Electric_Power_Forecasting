```python
import csv
import pandas as pd
import time

month = {1:31, 2:28, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}

for y in range(2013, 2017):
    
    for m in range(1,13):
        csv_origin = pd.read_csv("./Electirc_Data_CSV/"+str(y)+'_'+str(m)+'_'+"House1_Ch1_Electric_Data.csv", 
                  names = ['Time', 'Watt'], 
                  header = None, encoding='cp949')
        s = 0
        # 각 월의 일수만큼 
        for d in range(month[m]):
            d=d+1 # 밑에서 코드의 계산을 편리하게 하기 위한 1 더해주기. = 실제일과 맞추는 작업
            one_day_list_p1=[]
            one_day_list_d1=[]

            for n in range(s, s+24):
                date1 = csv_origin['Time'][n]
                power1 = csv_origin['Watt'][n]
                one_day_list_d1.append(date1)
                one_day_list_p1.append(power1)
                
            s=s+24
            
            
            if 'Null' in one_day_list_p1:
                N=0
                print(24*month[m])
                print('결손치 존재\n')
                for n in range(24):
                    try:
                        new_list=[]
                        new_list.append(one_day_list_d1[n])
                        new_list.append(int(n))

                        if one_day_list_p1[n] == 'Null':
                            N=n+24*d # N이 계속 커지다가 해당일의 한계일까지 더한 값이 되면 예외 조치를 해야함.(=월말)
                            power2 = csv_origin['Watt'][N] 
                            print('power2 >>', power2)

                            if power2 == 'Null':
                                N=n+24*(d+1)
                                print(N)
                                power3 = csv_origin['Watt'][N] 
                                print('power3 >>', power3)

                                if power3 == 'Null':
                                    N=n+24*(d+2)
                                    print(N)
                                    power4 = csv_origin['Watt'][N]
                                    print('power4 >>', power4)

                                    if power4 == 'Null':
                                        N=n+24*(d+3)
                                        print(N)
                                        power5 = csv_origin['Watt'][N]
                                        print('power5 >>', power5)

                                        if power5 == 'Null':
                                            N=n+24*(d+4)
                                            print(N)
                                            power6 = csv_origin['Watt'][N]
                                            print('power6 >>', power6)

                                            if power6 == 'Null':
                                                N=n+24*(d+5)
                                                print(N)
                                                power7 = csv_origin['Watt'][N]
                                                print('power7 >>', power7)

                                                if power7 == 'Null':
                                                    N=n+24*(d+6)
                                                    print(N)
                                                    power8 = csv_origin['Watt'][N]
                                                    print('power8 >>', power8)

                                                    if power8 == 'Null':
                                                        N=n+24*(d+7)
                                                        print(N)
                                                        power9 = csv_origin['Watt'][N]
                                                        print('power9 >>', power9)
                                                        new_list.append(power9)

                                                    else:
                                                        new_list.append(power8)
                                                else:
                                                    new_list.append(power7)
                                            else:
                                                new_list.append(power6)
                                        else:
                                            new_list.append(power5)
                                    else:
                                        new_list.append(power4)
                                else:
                                    new_list.append(power3)
                            else: 
                                new_list.append(power2)
                        else:
                            new_list.append(one_day_list_p1[n])

                        print(new_list)
                        print()
                        with open("./Electirc_Data_CSV_Null_Disposed/"+str(y)+'_'+str(m)+'_'+"House1_Ch1_Electric_Data_ND.csv", 'a', newline='') as file:
                                    writer = csv.writer(file)
                                    writer.writerow(new_list)
                    except KeyError:
                        print('드디어 찾았다')
                        time.sleep(30)
                
            else:
                print('정상일\n')
                
                for n in range(24):
                    new_list=[]
                    new_list.append(one_day_list_d1[n])
                    new_list.append(int(n))
                    new_list.append(one_day_list_p1[n])
                    print(new_list)
                    print()
                    with open("./Electirc_Data_CSV_Null_Disposed/"+str(y)+'_'+str(m)+'_'+"House1_Ch1_Electric_Data_ND.csv", 'a', newline='') as file:
                                writer = csv.writer(file)
                                writer.writerow(new_list)
                    
            print('\n'*4)
            
```

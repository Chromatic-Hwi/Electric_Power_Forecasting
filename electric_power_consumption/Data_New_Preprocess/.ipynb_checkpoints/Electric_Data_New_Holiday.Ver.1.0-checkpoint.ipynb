```python
import datetime
import pandas as pd
import numpy as np
import csv

def get_days(yyyy, mm, dd):
    days = ['월', '화', '수', '목', '금', '토', '일']
    return days[datetime.date(yyyy, mm, dd).weekday()]

"""
영국의 휴일로는 공통휴일로 선정.
새해 첫날, 성 금요일(=부활절 이틀 전 금요일), 노동절 휴일(5월 첫 월요일), 춘계 휴일(5월의 마지막 월요일), 크리스마스, 박싱데이(12/26)들과 대체휴일
"""

holidays = ['2013-01-01', '2013-03-29', '2013-05-06', '2013-05-27', '2013-12-25', '2013-12-26',
           '2014-01-01', '2014-04-18', '2014-05-05', '2014-05-26', '2014-12-25', '2014-12-26',
           '2015-01-01', '2015-04-03', '2015-05-04', '2015-05-25', '2015-12-25', '2015-12-26', '2015-12-28',
           '2016-01-01', '2016-03-25', '2016-05-02', '2016-05-30', '2016-12-25', '2016-12-26', '2016-12-27']

month = {1:31, 2:28, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}
day=''
holiday=''

for y in range(2013, 2017):
    for m in range(1,13):
        csv_for_days = pd.read_csv("./New_Electirc_Data_CSV/"+str(y)+'_'+str(m)+'_'+"2013_1_House1_Ch1_Electric_Data_NEW.csv", 
                                   names = ["Date", "Hour", "Minute" "Watt"], encoding='cp949')
        s = 0
        
        for d in range(month[m]):
            d=d+1
            
            yyyy=y
            
            mm=m
            if mm<=9:
                mm=str(mm)
                mm='0'+mm
                
            dd=d
            if dd<=9:
                dd=str(dd)
                dd='0'+dd
            
            ymd_for_graph = '{}-{}-{}'.format(yyyy,mm,dd) # 연속적인 그래프를 위한 새 변수
            mm=int(mm)
            dd=int(dd)
            day = get_days(yyyy, mm, dd) # 요일 변수
            
            one_day_list_p=[]
            one_day_list_d=[]
            
            # 조건부 표현식으로 간단하게 휴일 처리
            holiday='1' if ymd_for_graph in holidays or day=='토' or day=='일' else '0'
            
            for n in range(s, s+24):
                date = ymd
                power = csv_for_days['Watt'][n]
                one_day_list_d.append(date)
                one_day_list_p.append(power)
                
            for n in range(24):
                    new_list=[]
                    new_list.append(one_day_list_d[n])
                    new_list.append(ymd_for_graph)
                    new_list.append(day)
                    new_list.append(holiday)
                    new_list.append(int(n))
                    new_list.append(one_day_list_p[n])
                    
                    print(new_list)
                    print()
                    with open("./Electirc_Data_CSV_Final/"+str(y)+'/'+str(y)+'_'+str(m)+'_'+"House1_Ch1_Electric_Data_Final.csv", 'a', newline='') as file:
                                writer = csv.writer(file)
                                writer.writerow(new_list)
                                
            s=s+24
            
print('\n'*4,'------------------------------< 요일 입력, 휴일 처리 끝!! >------------------------------\n')  
```

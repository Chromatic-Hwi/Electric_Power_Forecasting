```python
"""
jump 를 하는 경우가 있기 때문에 이 경우를 구분 짓는 변수를 j로 선택

watt_list 에는 해당 시단위동안 모든 watt값들이 저장됨. 그중에서 max값을 추출하면 max_watt로 저장됨.

d4는 정상적으로 있어야 하는 일자인데 하루가 통째로 비는 경우를 처리하는 변수 D는 d4+1로 첫 생성
"""
```

```python
import csv

month = {1:31, 2:28, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}

# 1. - 시작 지점 지정
#i = 431368, 517000, 684157, 555910
i = 555910
print('Running!', '\n'*9)



# 2. - 전체 반복 While문 진입
while(i < 20284836):
    list1 = ch1_date.loc[i].values.tolist()
            
    date = list1[1]
    y1 = date.year
    M1 = date.month
    d1 = date.day
    H1 = date.hour # 이 H가 있어야 밑에 있는 하루 단위 반복이 돌 변수가 됨.
    n = 0 # 데이터 개수를 세기 위한 변수

        
        
    # 3. - 하루 단위 반복
    while(H1 < 24):
        
        H0 = 0 # H1은 데이터 원본에 존재하는 시각만 불러오지만, H0는 결손 여부와 무관하게 하루 전체를 표시. (0~23)
        
        # 3'. - 결손 데이터를 처리하고 하루를 넘기기 위한 반복문
        while(H0 < 24):
            
            watt_list=[]
            h = date.hour
            
            # 3.1. - 하루 시작이 0시부터 잘 이뤄지는 경우
            if(H1 == H0):
                
                # 3.1'. - h시와 h+1 시 사이 반복
                while(h < h+1):
                    n=n+1
                    list2 = ch1_date.loc[i].values.tolist()

                    date = list2[1]
                    watt = list2[2]

                    y = date.year
                    M = date.month
                    d = date.day
                    h = date.hour
                    m = date.minute
                    s = date.second

                    print()

                    j=h-H1

                    # 3.1.1. - 시간정보가 제때에 맞는 경우, 계속 반복하게 되는 동작
                    if(j==0):
                        watt_list.append(watt)

                        # 확인용 각 변수 출력
                        total_date = '{0}년 {1}월 {2}일 - {3}시 {4}분 {5}초'.format(y, M, d, h, m, s)
                        print(total_date)
                        print('d1 >>', d1, 'd >>', d, 'H1 >>', H1, 'h >>', h, '  //  ','i >>', i)
                        print('칼럼 출력 >>', list2)

                        i = i+1


                    # 3.1.2. - 시단위가 바로 다음으로 넘어가는 경우
                    elif (j==1):
                        print()
                        
                        print('시 단위 변경! 루프 종료 및 맥스값 추출!')
                        print('========================================< 3.1.2. - 1시간치 끝!! >==================================================')
                        print()
                        
                        print('d1 >>', d1, 'd >>', d, 'H1 >>', H1, 'H0 >>', H0, 'h >>', h, '  //  ','i >>', i)
                        print('1시간 동안 데이터 수 >>', n)
                        print()

                        max_watt = max(watt_list)
                        print(H1, '시까지의 최대값 >>', max_watt)

                        # 1시간 동안이니 해당 시각의 정보를 시 단위까지로 해서 새로 저장
                        hourly_date2 = '{0}년 {1}월 {2}일 - {3}시'.format(y, M, d, H1)
                        hourly_date5 = '{0} {1} {2}'.format(y, M, d) # 밑의 경우를 위한 별도 저장
                        hourly_list = []
                        hourly_list.append(hourly_date2)
                        hourly_list.append(max_watt)

                        # 데이터 저장 
                        with open("./"+str(y)+'_'+str(M)+'_'+"House1_Ch1_Electric_Data.csv", 'a', newline='') as file:
                            writer = csv.writer(file)
                            writer.writerow(hourly_list)       
                        print('*csv 저장 완료! - 3.1.2.')

                        H1 = H1+1 # 저장했으니 다음 시각으로 넘어가기 위한 +1
                        H0 = H0+1
                        print('\n'*4)

                        break

                    # 3.1.3. - 1 시간치 정리가 끝났는데 일 단위가 다음 날로 넘어가는 경우
                    elif(d1+1==d and h ==0 and H1 ==23):
                        print('========================================< 3.1.3. - 1시간치 끝!! >==================================================')
                        print()
                        
                        print('d1 >>', d1, 'd >>', d, 'H0 >>', H0, 'H1 >>', H1, 'h >>', h, '  //  ','i >>', i)
                        print('1시간 동안 데이터 수 >>', n)
                        print()
                        
                        d=d-1
                        h=h+24

                        max_watt = max(watt_list)
                        print(H1, '시까지의 최대값 >>', max_watt)
                        
                        hourly_date3 = '{0}년 {1}월 {2}일 - {3}시'.format(y, M, d, H1)
                        hourly_list = []
                        hourly_list.append(hourly_date3)
                        hourly_list.append(max_watt)

                        # 데이터 저장 
                        with open("./"+str(y)+'_'+str(M)+'_'+"House1_Ch1_Electric_Data.csv", 'a', newline='') as file:
                            writer = csv.writer(file)
                            writer.writerow(hourly_list)       
                        print('*csv 저장 완료! - 3.1.3.')

                        H1 = H1+1 # 저장했으니 다음 시각으로 넘어가기 위한 +1
                        H0 = H0+1
                        print('\n'*4)
                        
                        print('========================================< 하루 끝!! >==================================================')
                        print('d1 >>', d1, 'd >>', d, 'H0 >>', H0, 'H1 >>', H1, 'h >>', h, '  //  ','i >>', i)
                        print('\n'*4)
                        
                        break

                    
                    # 3.1.4. - 한시간의 데이터가 600개를 넘지 않기 때문에 별도의 경우로 설정. 결손된 데이터를 처라하기 위한 케이스
                    elif (n>600):
                        list(hourly_date5)
                        listed = hourly_date5.split()
                        y4 = listed[0]
                        M4 = listed[1]
                        d4 = listed[2]


                        # 일단 해당 시각 존재하는 만큼은 정리해서 맥스값을 추출
                        print()
                        
                        print('점프! 시 단위 변경! 루프 종료 및 맥스값 추출!')
                        print('========================================< 3.1.4. - 1시간치 끝!! >==================================================')
                        print()

                        print('d1 >>', d1, 'd >>', d, 'H0 >>', H0, 'H1 >>', H1, 'h >>', h, '  //  ','i >>', i)

                        print('1시간 동안 데이터 수 >>', n)
                        print()

                        # 이 지점이 1시간 간격내에서 구해진 watt 값들 중 max를 뽑는 지점.
                        max_watt = max(watt_list)
                        print(H1, '시까지의 최대값 >>', max_watt)

                        # 1시간 동안이니 해당 시각의 정보를 시 단위까지로 해서 새로 저장
                        hourly_date = '{0}년 {1}월 {2}일 - {3}시'.format(y4, M4, d4, H1)
                        hourly_list = []
                        hourly_list.append(hourly_date)
                        hourly_list.append(max_watt)

                        # 데이터 저장 
                        with open("./"+str(y4)+'_'+str(M4)+'_'+"House1_Ch1_Electric_Data.csv", 'a', newline='') as file:
                            writer = csv.writer(file)
                            writer.writerow(hourly_list)       
                        print('*csv 저장 완료! - 3.1.4.') 
                        print('\n'*4)
                        
                        # 3.1.4.1. - 같은 날짜 안인 경우
                        if(d1==d):
                            jc=j
                            
                            while(jc>1):
                                h=h+1
                                # 점프한 시각 이전까지 Null로 입력 
                                hourly_date = '{0}년 {1}월 {2}일 - {3}시'.format(y4, M4, d4, h-j)
                                hourly_list = []
                                hourly_list.append(hourly_date)
                                hourly_list.append('Null')

                                # 데이터 저장 
                                with open("./"+str(y4)+'_'+str(M4)+'_'+"House1_Ch1_Electric_Data.csv", 'a', newline='') as file:
                                    writer = csv.writer(file)
                                    writer.writerow(hourly_list)  

                                print(hourly_list, jc)
                                print('*csv 저장 완료! - 3.1.4.1.')
                                print()
                                jc=jc-1
                                H1 = H1+1
                                H0 = H0+1
                                
                            H1 = H1+1
                            H0 = H0+1
                            
                        #break
                                
                        # 3.1.4.2. - 결손 데이터 지속 중 일자가 달라지는 경우. 해당 일의 잔여시를 처리하는 코드
                        if(d1!=d):
            
                            # 건너뛴 횟수만큼 빈 시간대를 Null로 저장
                            while(H1<23):
                                # 점프한 시각 이전까지 Null로 입력 
                                hourly_date = '{0}년 {1}월 {2}일 - {3}시'.format(y4, M4, d4, H1+1)
                                hourly_list = []
                                hourly_list.append(hourly_date)
                                hourly_list.append('Null')

                                # 데이터 저장 
                                with open("./"+str(y4)+'_'+str(M4)+'_'+"House1_Ch1_Electric_Data.csv", 'a', newline='') as file:
                                    writer = csv.writer(file)
                                    writer.writerow(hourly_list)  

                                print(hourly_list)
                                print('*csv 저장 완료! - 3.1.4.2.')
                                print()
                                H1 = H1+1
                                H0 = H0+1
                                
                            H1 = H1+1
                            H0 = H0+1
                            
                            # 3.1.4.2.1. - 일자를 봤더니 다음 날과 딱 하루 차이가 나지 않음 = 하루가 통째로 빈 경우. 그것을 처리하는 코드
                            if(int(d4)+1 != d):
                                
                                # 해당 달 내에서 반복
                                D = int(d4)+1 
                                
                                # 3.1.4.2.1.1. - 월 중간에서 데이터가 없는 날이 있는 경우
                                if(int(M) == int(M4)):
                                    while(D<d):
                                        for x in range (0,24):
                                            hourly_date = '{0}년 {1}월 {2}일 - {3}시'.format(y4, M4, D, x)
                                            hourly_list = []
                                            hourly_list.append(hourly_date)
                                            hourly_list.append('Null')

                                            # 데이터 저장 
                                            with open("./"+str(y4)+'_'+str(M4)+'_'+"House1_Ch1_Electric_Data.csv", 'a', newline='') as file:
                                                writer = csv.writer(file)
                                                writer.writerow(hourly_list)  

                                            print(hourly_list)
                                            print('*csv 저장 완료! - 월중 경우')
                                            print()
                                        D=D+1
                                
                                # 3.1.4.2.1.2. - 월말에서 다음 달까지 데이터가 없는 경우
                                if(int(M) != int(M4)):
                                    while(D<=month[int(M4)]):
                                        for x in range (0,24):
                                            hourly_date = '{0}년 {1}월 {2}일 - {3}시'.format(y4, M4, D, x)
                                            hourly_list = []
                                            hourly_list.append(hourly_date)
                                            hourly_list.append('Null')

                                            # 데이터 저장 
                                            with open("./"+str(y4)+'_'+str(M4)+'_'+"House1_Ch1_Electric_Data.csv", 'a', newline='') as file:
                                                writer = csv.writer(file)
                                                writer.writerow(hourly_list)  

                                            print(hourly_list)
                                            print('*csv 저장 완료! - 월말 경우')
                                            print()
                                        D=D+1
                                        
                                    D2 = 1
                                    while(D2 < d):
                                        for x in range (0,24):
                                            hourly_date = '{0}년 {1}월 {2}일 - {3}시'.format(y4, M, D2, x)
                                            hourly_list = []
                                            hourly_list.append(hourly_date)
                                            hourly_list.append('Null')

                                            # 데이터 저장 
                                            with open("./"+str(y4)+'_'+str(M)+'_'+"House1_Ch1_Electric_Data.csv", 'a', newline='') as file:
                                                writer = csv.writer(file)
                                                writer.writerow(hourly_list)  

                                            print(hourly_list)
                                            print('*csv 저장 완료! - 3.1.4.2.2.')
                                            print()
                                        D2 = D2+1

                        break

                        
            # 3.2. - 하루의 시작이 0시부터 순차적으로 진행되는 것이 아닌 결손되어 있는 경우
            if(H1 != H0):

                list3 = ch1_date.loc[i].values.tolist()

                date = list3[1]
                watt = list3[2]

                y = date.year
                M = date.month
                d = date.day
                h = date.hour
                m = date.minute
                s = date.second

                print()

                # jump 를 하는 경우가 있기 때문에 이 경우를 구분 짓는 변수를 j로 선택
                j=h-H1
                    
                print('시작 시각 불일치!!')
                print('H0 >>', H0, 'H1 >>', H1, 'h >>', h, 'j >>', j)
                

                hourly_date3121 = '{0}년 {1}월 {2}일 - {3}시'.format(y, M, d, H0)
                hourly_list3121 = []
                hourly_list3121.append(hourly_date3121)
                print(hourly_list3121)
                hourly_list3121.append('Null')

                # 데이터 저장 
                with open("./"+str(y)+'_'+str(M)+'_'+"House1_Ch1_Electric_Data.csv", 'a', newline='') as file:
                    writer = csv.writer(file)
                    writer.writerow(hourly_list3121)       
                print('*csv 저장 완료! - 3.2.')
                print()
                print('========================================< 3.2. - 1시간치 끝!! >==================================================')
                
                
                h = h+1
                H0 = H0+1
                print()
                
                # 만약 하루의 시작부터 계속해서 데이터가 없다면 23 즉, 다음날로 넘기기 위함.
                if (H0 ==23):
                    break
                    
        # 3. 의 끝 지점
        print('종료 지점 확인용','\n'*9)
        break

        
    
        
# 전체 While문 종료 지점
```

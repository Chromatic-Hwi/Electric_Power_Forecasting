```python
def arc (year, month, day):
        h = 0 # 여기서는 하루를 30분 단위로 쪼갠 온도 개수, 데이터의 개수를 h로 설정해서 반복시켰는데, 48이 안되는 경우가 있기 때문에
        while(h<48):
            h=h+1
            try:
                clk = driver.find_element(By.XPATH,'//*[@id="inner-content"]/div[2]/div[1]/div[5]/div[1]/div/lib-city-history-observation/div/div[2]/table/tbody/tr['+str(h)+']/td[1]/span')
                clk_str = clk.text
                clk_str.split()
                temp_list = [int(year), int(month), int(day)]
                temp_list.append(clk_str)
                temp_list.append(int(h))
                temp = driver.find_element(By.XPATH,'//*[@id="inner-content"]/div[2]/div[1]/div[5]/div[1]/div/lib-city-history-observation/div/div[2]/table/tbody/tr['+str(h)+']/td[2]')
                temp_str = temp.text
                temp_list.append(temp_str)
                time.sleep(0.1)
                print(temp_list) # 확인용 print
                # 여기가 시간 루프 종료 지점
                    
                # csv 추가 및 저장
                with open("./"+str(year)+'_'+str(month)+'_'+"London_Temperature.csv", 'a', newline='') as file:
                    writer = csv.writer(file)
                    writer.writerow(temp_list)
                        
                    
            except NoSuchElementException:
                print('데이터 48개 미달')
                break


        # 그러면 여기가 하루 종료 지점?
        print('하루끝')
        print()
        time.sleep(3)
```
